```python
for col in categorical_cols:
    df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col].astype(str))
```

```python
df_con_nulos = df_encoded.copy()
df_sin_nulos = df_encoded.copy()
df_sin_nulos = df_sin_nulos.dropna()
```

```python
df_sin_nulos.isnull().sum()
```

```python
def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# Aplicar máscara para generar df_con_nulos con valores faltantes
mask = np.random.rand(*df_sin_nulos.shape) < 0.2
df_con_nulos = df_sin_nulos.mask(mask)

mse = []
k_range = range(1, 21)

kf = KFold(n_splits=5, shuffle=True, random_state=357)

for k in k_range:
    fold_errors = []
    imputer = KNNImputer(n_neighbors=k)

    for train_index, test_index in kf.split(df_con_nulos):
        # Separar conjuntos de entrenamiento y prueba
        df_titanic_train = df_con_nulos.iloc[train_index]
        df_titanic_test = df_con_nulos.iloc[test_index]

        # Imputar valores faltantes usando solo los datos de entrenamiento
        df_titanic_imputed = imputer.fit_transform(df_titanic_train)
        
        # Predecir para los datos de prueba
        df_titanic_test_imputed = imputer.transform(df_titanic_test)

        # Calcular el RMSE entre el conjunto de prueba original y el imputado
        mse_fold = root_mean_squared_error(
            df_sin_nulos.iloc[test_index].values, df_titanic_test_imputed
        )
        fold_errors.append(mse_fold)

    mse_mean = np.mean(fold_errors)
    mse.append(mse_mean)
    print(f'k={k} - mse={mse_mean}')

# Graficar los resultados
plt.plot(k_range, mse, marker='o')
plt.xlabel('Edades K')
plt.ylabel('Root Mean Squared Error (RMSE)')
plt.title('MSE vs Edades K')
plt.show()
```
